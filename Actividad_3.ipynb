## Ejercicio 3

Define una función dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano

```python
class Numero:
    def __init__(self, numero):
        self.romano = self.convertir_a_romano(numero)
        self.normal = numero

    def convertir_a_romano(self, numero):
        simbolos_romanos = {
            1000: 'M',
            900: 'CM',
            500: 'D',
            400: 'CD',
            100: 'C',
            90: 'XC',
            50: 'L',
            40: 'XL',
            10: 'X',
            9: 'IX',
            5: 'V',
            4: 'IV',
            1: 'I'
        }
        romano = ''
        for valor, simbolo in simbolos_romanos.items():
            while numero >= valor:
                romano += simbolo
                numero -= valor

        return romano
    
    def imprime(self):
        print("Número en formato romano:", self.romano)
        print("Número normal:", self.normal)

    def suma_romano(self, otro_romano):
        if not isinstance(otro_romano, str):
            raise ValueError("El parámetro debe ser una cadena de caracteres.")
        if not self.es_numero_romano_valido(otro_romano):
            raise ValueError("La cadena no corresponde a un número romano válido.")
        otro_numero = self.convertir_a_numero(otro_romano)
        self.normal += otro_numero
        self.romano = self.convertir_a_romano(self.normal)
    
    def convertir_a_numero(self, romano):
        simbolos_romanos = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        numero = 0
        prev_valor = 0
        for caracter in reversed(romano):
            valor = simbolos_romanos.get(caracter, 0)
            if valor < prev_valor:
                numero -= valor
            else:
                numero += valor

            prev_valor = valor

        return numero
    def es_numero_romano_valido(self, cadena):
        """Verifica si la cadena corresponde a un número romano válido."""
        patrones_validos = ["I", "IV", "V", "IX", "X", "XL", "L", "XC", "C", "CD", "D", "CM", "M"]

        if not all(caracter in "IVXLCDM" for caracter in cadena):
            raise ValueError("La cadena contiene caracteres inválidos para un número romano.")

        for patron in patrones_validos:
            if cadena.startswith(patron):
                return True

        raise ValueError("La cadena no corresponde a un número romano válido.")
```

## Ejercicio 5

En la clase MejorNumero, crea un nuevo método que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). Los posibles errores se tendrán que gestionar con excepciones para mostrar un mensaje y seguir ejecutando el siguiente número.

```python
class MejorNumero(Numero):
    def __init__(self, numero):
        super().__init__(numero)

    def resta_romano(self, otro_romano):
        """Resta otro número romano al número actual."""
        if not isinstance(otro_romano, str):
            raise ValueError("El parámetro debe ser una cadena de caracteres.")

        if not self.es_numero_romano_valido(otro_romano):
            raise ValueError("La cadena no corresponde a un número romano válido.")

        otro_numero = self.convertir_a_numero(otro_romano)
        self.normal -= otro_numero
        self.romano = self.convertir_a_romano(self.normal)

    def multiplica_romano(self, otro_romano):
        """Multiplica el número actual por otro número romano."""
        if not isinstance(otro_romano, str):
            raise ValueError("El parámetro debe ser una cadena de caracteres.")

        if not self.es_numero_romano_valido(otro_romano):
            raise ValueError("La cadena no corresponde a un número romano válido.")

        otro_numero = self.convertir_a_numero(otro_romano)
        self.normal *= otro_numero
        self.romano = self.convertir_a_romano(self.normal)

    def suma_lista_romanos(self, lista_romanos):
        """Suma una lista de números romanos al número actual."""
        for romano in lista_romanos:
            try:
                self.suma_romano(romano)
            except ValueError as e:
                print(f"Error: {e}")
```
